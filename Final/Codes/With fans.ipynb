```python
#load the data
import pandas as pd

data = pd.read_csv("results.csv", encoding = 'ANSI')
```

```python
#put the data in a dataframe
dataset = pd.DataFrame(data)

#filter the dataframe and get 2018-2019 season's results only
data2018 = dataset[dataset['Season'] == "2018-19"] [['HomeTeam','AwayTeam', 'FTHG', 'FTAG', 'HS','AS', 'HST','AST']]
```

```python
#create a dataframe with it
withfans = pd.DataFrame(data2018)
```

```python
x = withfans
```

```python
#Make a new column as Goal Difference (Home Goals - Away Goals)
x['Goal Difference'] = x['FTHG'] - x['FTAG']

#Record the result by running a function
def findwinner(value):
    if value > 0:
        return("Home Wins")
    elif value < 0:
        return ("Away Wins")
    else:
        return("Draw")

#Apply the function and add a new column 'Result'
x['Result'] = x['Goal Difference'].apply(findwinner)

x.rename(columns={'FTHG': 'HomeGoal', 'FTAG': 'AwayGoal','HS':'HomeShot', 'AS':'AwayShot',
                  'HST':'HomeShotOnTarget', 'AST': 'AwayShotOnTarget','Goal Difference':'GoalDifference'}, inplace=True)

#Calculate the shot qulaity(Shot on Target per shot) of home and away teams respectively
x['HomeShotQuality'] = (x['HomeShotOnTarget'] / x['HomeShot'])*100
x['AwayShotQuality'] = (x['AwayShotOnTarget'] / x['AwayShot'])*100

#Calculate the shot conversion(Goals per Shot) of home and away teams respectively
x['HomeShotConversion'] = (x['HomeGoal'] / x['HomeShot'])*100
x['AwayShotConversion'] = (x['AwayGoal'] / x['AwayShot'])*100
```
