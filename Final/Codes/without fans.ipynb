```python
import pandas as pd
```

```python
#import the dataset
data = pd.read_csv("results.csv")

#put the dataset into a dataframe
dataset = pd.DataFrame(data)

#filter the dataset and get the data from 2020-2021's season
data2021 = dataset[dataset['Season'] == "2020-21"] [['HomeTeam','AwayTeam', 'FTHG', 'FTAG', 'HS','AS', 'HST','AST']]
```

```python
import pandas as pd
import math
```

```python
#put the filtered data into a dataframe
nofans2021 = pd.DataFrame(data2021)
```

```python
#delete games the have fans (fans were allowed to watch the last two gameweekes in person)
for i in range(10784,10804):
    nofans2021 = nofans2021.drop([i])
    i = i+1
```

```python
nofans2021
```

```python
x = nofans2021
```

```python
#Make a new column as Goal Difference (Home Goals - Away Goals)
x['Goal Difference'] = x['FTHG'] - x['FTAG']

#Record the result by running a function
def findwinner(value):
    if value > 0:
        return("Home Wins")
    elif value < 0:
        return ("Away Wins")
    else:
        return("Draw")
    
#Apply the function and add a new column 'Result'
x['Result'] = x['Goal Difference'].apply(findwinner)

x.rename(columns={'FTHG': 'HomeGoal', 'FTAG': 'AwayGoal','HS':'HomeShot', 'AS':'AwayShot',
                  'HST':'HomeShotOnTarget', 'AST': 'AwayShotOnTarget','Goal Difference':'GoalDifference'}, inplace=True)

#Calculate the shot qulaity(Shot on Target per shots) of home and away teams respectively
x['HomeShotQuality'] = (x['HomeShotOnTarget'] / x['HomeShot'])*100
x['AwayShotQuality'] = (x['AwayShotOnTarget'] / x['AwayShot'])*100

#Calculate the shot conversion(Goal per shot) of home and away teams respectively
x['HomeShotConversion'] = (x['HomeGoal'] / x['HomeShot'])*100
x['AwayShotConversion'] = (x['AwayGoal'] / x['AwayShot'])*100

x
```
