```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#Import the datasets that have been processed : match results with fans and without fans
withfans = pd.read_csv("with_fans.csv")
nofans = pd.read_csv("without_fans.csv")
```

```python
#home goal average
hg1 = withfans['HomeGoal'].mean()
hg2 = nofans['HomeGoal'].mean()

#away goal average
ag1 = withfans['AwayGoal'].mean()
ag2 = nofans['AwayGoal'].mean()

#home shot average
hs1 = withfans['HomeShot'].mean()
hs2 = nofans['HomeShot'].mean()

#away shot average
as1 = withfans['AwayShot'].mean()
as2 = nofans['AwayShot'].mean()

#goal difference average 
gd1 = withfans['GoalDifference'].mean()
gd2 = nofans['GoalDifference'].mean()

#home shot quality average
hsq1 = withfans['HomeShotQuality'].mean()
hsq2 = nofans['HomeShotQuality'].mean()

#away shot quality average
asq1 = withfans['AwayShotQuality'].mean()
asq2 = nofans['AwayShotQuality'].mean()

#homeconversion average
hc1 = withfans['HomeShotConversion'].mean()
hc2 = nofans['HomeShotConversion'].mean()

#awayconversion average
ac1 = withfans['AwayShotConversion'].mean()
ac2 = nofans['AwayShotConversion'].mean()
```

```python
#Count the number of 'Home Wins' from the 'Result' column and calculate the win rate for home teams
homewins1 = withfans[withfans['Result'] == 'Home Wins']
homewr1 = ((len(homewins1) / 380)*100) 
homewins2 = nofans[nofans['Result'] == 'Home Wins']
homewr2 = ((len(homewins2) / 380)*100) 

#Count the number of 'Away Wins' from the 'Result' column and calculate the win rate for away teams
awaywins1 = withfans[withfans['Result'] == 'Away Wins']
awaywr1 = ((len(awaywins1) / 380)*100) 
awaywins2 = nofans[nofans['Result'] == 'Away Wins']
awaywr2 = ((len(awaywins2) / 380)*100) 

#Count the number of 'Draw' from the 'Result' column and calculate rate of tied games
draws1 = withfans[withfans['Result'] == 'Draw']
draws1 = ((len(draws1) / 380)*100) 
draws2 = nofans[nofans['Result'] == 'Draw']
draws2 = ((len(draws2) / 380)*100) 
```

```python
#Create a bar chart showing the average shots by Home Team and Away Team
#Also showing the comparison between withfans and withoutfans
barchart1 = pd.DataFrame({"With Fans":[hs1,as1],"Without Fans":[hs2,as2]}, index = ["Home", "Away"])

x = barchart1.plot.bar(figsize=(15,8), colormap = 'Paired',rot=0, width = 0.3)
for p in x.containers:
    x.bar_label(p, fmt='%.2f', label_type='center')

plt.ylabel('Average number')
plt.title('Average shots')
plt.savefig('Shots.jpg')
```

```python
#Create another bar chart showing the shot quality of Home Team and Away Team
#Also showing the comparison between withfans and withoutfans
barchart2 = pd.DataFrame({"With Fans":[hsq1,asq1],"Without Fans":[hsq2,asq2]}, index = ["Home", "Away"])

y = barchart2.plot(kind="bar",figsize=(15,8), colormap = 'YlOrRd',rot=0, width = 0.3)
for p in y.containers:
    y.bar_label(p, fmt='%.2f', label_type='center')

plt.ylabel('Average(%)')
plt.title('Shot Quality')
plt.savefig('Shots_Quality.jpg')
plt.show()
```

```python
#Create another bar chart showing the shot conversion of Home Team and Away Team
#Also showing the comparison between withfans and withoutfans
barchart3 = pd.DataFrame({"With Fans":[hc1,ac1],"Without Fans":[hc2,ac2]}, index = ["Home", "Away"])

plot1 = barchart3.plot.bar(figsize=(15,8), colormap = 'cool',rot = 0, width = 0.3)
for p in plot1.containers:
    plot1.bar_label(p, fmt='%.2f', label_type='center')

plt.ylabel('Average(%)')
plt.title('Shot Conversion')
plt.savefig('Shot_Conversion.jpg')
plt.show()
```

```python
#Create another bar chart showing the number of goals scored by Home Team and Away Team
#Also showing the comparison between withfans and withoutfans
barhchart1 = pd.DataFrame({"With Fans":[hg1,ag1],"Without Fans":[hg2,ag2]}, index = ["Home", "Away"])

plot2 = barhchart1.plot.barh(figsize=(15,8), colormap = 'plasma',rot = 0, width = 0.3)
for p in plot2.containers:
    plot2.bar_label(p, fmt='%.2f', label_type='edge')

plt.xlabel('Average Number')
plt.title('Goals Scored')
plt.savefig('Goals.jpg')
plt.show()
```

```python
#Putting the goal difference from withfans matches and withoutfans matches
data = [gd1,gd2]
df = pd.DataFrame(data,columns = ['Goal Difference'], index = ['With Fans','Without Fans'])

#Create a horizontal bar chart showing the goal difference from matches withfans and withoutfans
x = df.plot.barh(figsize=(10,8), colormap = 'summer',rot = 0, width = 0.3)
for p in x.containers:
    x.bar_label(p, fmt='%.2f', label_type='edge')

plt.xlabel('Average Number')
plt.title('Goal Difference')
plt.savefig('Goal_Difference.jpg')
plt.show()
```

```python
#Create an array consits of the home team win rate, away team win rate, and rate of tied games before covid restrictions
x = [homewr1,awaywr1,draws1]

#Create a pie chart showing the distribution 
colours1 = ['#DC143C','#3D59AB','#006400']
withfanspie=plt.pie(x,labels=['Home','Away','Draw'],colors=colours1,autopct='%.1f%%',startangle=90, 
                    wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},textprops={'size': 'x-large'})

plt.title('Win rate with Fans')
plt.show()
```

```python
#Create an array consits of the home team win rate, away team win rate, and rate of tied games after covid restrictions
y = [homewr2,awaywr2,draws2]
colours2 = ['#DC143C','#3D59AB','#006400']
withoutfanspie=plt.pie(y,labels=['Home','Away','Draw'],colors=colours2,autopct='%.1f%%',startangle=90, 
                       wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},textprops={'size': 'x-large'})

plt.title('Win rate without Fans')
plt.show()
```

```python
#Putting two charts side by side to show the comparison
fig, ax = plt.subplots(1,2,figsize=(7,7))
fig.tight_layout()
ax[0].pie(x,labels=['Home','Away','Draw'],colors=colours1,autopct='%.1f%%',startangle=90, wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})
ax[0].set_title('Win rate with Fans')



ax[1].pie(y,labels=['Home','Away','Draw'],colors=colours2,autopct='%.1f%%',startangle=90, wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})
ax[1].set_title('Win rate without Fans')

plt.savefig('Winrate.jpg')
```
